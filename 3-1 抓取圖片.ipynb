### 抓取圖片

抓取網路上的圖形，並存成檔案。

例如：下載底下圖片

url = "http://i.epochtimes.com/assets/uploads/2018/08/PO_X95711-600x400.jpg"

或

url = "https://i2.kknews.cc/SIG=19mlbso/5r8700046p0qr32258q1.jpg"

```python
import requests
from PIL import Image
from io import BytesIO
```

```python
url = "http://i.epochtimes.com/assets/uploads/2018/08/PO_X95711-600x400.jpg"
```

```python
# 抓取圖片
r = requests.get(url)
```

```python
# 存成檔案
with open("d:\\相簿\\lin.jpg", "wb") as f:
    f.write(r.content)
```

### 直接抓圖片做影像處理

```python
lin = Image.open(BytesIO(r.content))
```

```python
lin
```

```python
# 分離色層
r, g, b = lin.split()
# 改變色層順序，再合併成彩色圖形
lin = Image.merge("RGB", (b, g, r))
lin
```
